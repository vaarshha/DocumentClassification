```python
import numpy as np 
import pandas as pd 
import seaborn as sns
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

```python
label_df = pd.read_csv("../input/datathonindoml-2022/train_labels.csv")
```

```python
sorted(label_df["label"].unique())
```

```python
classes = {
    0:"letter",1:"memo",2:"email",3:"filefolder",
    4:"form",5:"handwritten",6:"invoice",7:"advertisement",
    8:"budget",9:"news article",10:"presentation",11:"scientific publication",
    12:"questionnaire",13:"resume",14:"scientific report",15:"specification"
}
```

```python
label_df['label'].value_counts()
```

```python
# Seeing the frequency distribution
plt.rcParams["figure.figsize"] = [7.50, 3.50]
fig, ax = plt.subplots()
label_df['label'].value_counts().plot(ax=ax, kind='bar', xlabel='Classes', ylabel="Count")

plt.show()
```

```python
# Plotting an example of each type of document
for i in range(16):
    v = label_df.loc[label_df['label'] == i].values[0][0]
    img = cv2.imread('../input/datathonindoml-2022/train/train/'+str(v)+".tif")
    print(classes[i])
    plt.imshow(img)
    temp=cv2.resize(img,(512,512))
    plt.imshow(temp)
    plt.show()
```
